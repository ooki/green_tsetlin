```python
import os 
import pickle

import numpy as np
import pandas as pd

import umap
import umap.plot
```

```python


in_file = "results/mnist_dense_results_101073879878206.pkl"
with open(in_file, "rb") as fp:
    data = pickle.load(fp)
    
print("cpu:", data["cpu"]["brand_raw"])
print("physical_cores: {}".format(data["physical_cores"]))
print("logical_cores: {}".format(data["logical_cores"]))
print("#trials:", len(data["trials"]))

trials = data["trials"]
print(trials[0])

df = pd.DataFrame.from_records(trials)
df
    
```

```python

from sklearn import linear_model
from sklearn import svm

sv = svm.SVR()
ols = linear_model.LinearRegression()
ridge = linear_model.Ridge(alpha=0.1)
enet = linear_model.ElasticNet(random_state=0)


# TODO: normalize the dataset (also try to put in #ta = n_clause * n_literals * 2 as a feature)

for reg, name in [(ols, "OLS"), (ridge, "Ridge"), (enet, "enet"), (sv, "svm")]:
    xs = df[["n_literals", "n_examples",	"n_clauses", "n_cbs", "n_jobs"]]
    y = df["time"]
    reg.fit(xs, y)
    print("{}: {:.3f}".format(name, reg.score(xs,y)))
```
